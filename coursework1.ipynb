#### Find the quarterly value of new cases

 1. Calculate the new cases in a month
 2. Sum up the new cases of each month
 3. Append the sum of the new cases in a list
 4. Add the sum in the list for the total value of new cases in a quarter year (Every quarter being 3 months)
 5. Append the final total value into a dictionary
 6. Repeat the above 5 stops for 4 quarters.
 6. Create the dataframe for the dictionary
 7. Reset the index of the dataframe.

```python
# Function to calculate the total covid new cases for the month.
def sumMonthNewCases(x,dt_string):
    month_df = covid_daily_sg_2020[covid_daily_sg_2020['date'].str.contains(dt_string)]
    new_cases = month_df['new_cases']
    total_cases_per_month = new_cases.sum()
    return total_cases_per_month
```

```python
import datetime as dt

#dictonary to store data of new covid cases every quarter
covid_quarterly_sg_2020_dict = {}
```

```python
# for loop for Q1 of the year
list_for_q1 = []
for x in range(1,4):
    dt_string = "{}-0{}".format(2020,x)
    list_for_q1.append(sumMonthNewCases(x,dt_string))

# add sum of total new cases together for total of new cases every quater
total_cases_for_q1 = 0    
for y in list_for_q1:
    total_cases_for_q1 += y
    
if "Q1" not in covid_quarterly_sg_2020_dict:
    covid_quarterly_sg_2020_dict["2020-Q1"] = total_cases_for_q1
```

```python
# for loop for Q2 of the year
list_for_q2 = []
for x in range(4,7):
    dt_string = "{}-0{}".format(2020,x)
    list_for_q2.append(sumMonthNewCases(x,dt_string))
    
total_cases_for_q2 = 0    
for y in list_for_q2:
    total_cases_for_q2 += y

if "Q2" not in covid_quarterly_sg_2020_dict:
    covid_quarterly_sg_2020_dict["2020-Q2"] = total_cases_for_q2
```

```python
# for loop for Q3 of the year
list_for_q3 = []
for x in range(7,10):
    dt_string = "{}-0{}".format(2020,x)
    list_for_q3.append(sumMonthNewCases(x,dt_string))
    
total_cases_for_q3 = 0    
for y in list_for_q3:
    total_cases_for_q3 += y
    
if "Q3" not in covid_quarterly_sg_2020_dict:
    covid_quarterly_sg_2020_dict["2020-Q3"] = total_cases_for_q3
```

```python
# for loop for Q4 of the year
list_for_q4 = []
for x in range(10,13):
    dt_string = "{}-{}".format(2020,x)
    list_for_q4.append(sumMonthNewCases(x,dt_string))
    
total_cases_for_q4 = 0    
for y in list_for_q4:
    total_cases_for_q4 += y
    
if "Q4" not in covid_quarterly_sg_2020_dict:
    covid_quarterly_sg_2020_dict["2020-Q4"] = total_cases_for_q4
```

```python
# Convert the collected dictionary data into DataFrame
covid_quarterly_sg_2020 = pd.DataFrame.from_dict(covid_quarterly_sg_2020_dict,orient='index',columns=['quarterly_total_new_covid_cases'])
covid_quarterly_sg_2020['quarter'] = covid_quarterly_sg_2020.index
covid_quarterly_sg_2020 = covid_quarterly_sg_2020[['quarter','quarterly_total_new_covid_cases']]

# Reset the index of the DataFrame
covid_cleaned = covid_quarterly_sg_2020.reset_index(drop=True)

covid_cleaned
```
